```python
import cv2
import pytesseract

# Path to the video file
video_path = r'D:\SRM\3rd_Year\6th_Semester\Research\Video_Retriveal_System\Data\NPTEL\Video\01.mp4'

# Initialize the video capture object
cap = cv2.VideoCapture(video_path)

# Initialize variables for adaptive thresholding
prev_frame_gray = None
stable_count = 0

# Initialize pytesseract OCR configuration
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Initialize a string variable to accumulate text content
all_text_content = ""

# Process video frames
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convert frame to grayscale
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if prev_frame_gray is not None:
        # Calculate absolute difference between consecutive frames
        abs_diff = cv2.absdiff(frame_gray, prev_frame_gray)

        # Apply adaptive mean thresholding
        _, thresh = cv2.threshold(abs_diff, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Count non-zero pixels in the thresholded image
        non_zero_pixels = cv2.countNonZero(thresh)

        # Determine if the frame is stable (i.e., contains little variation)
        if non_zero_pixels < 0.5 * frame_gray.size:
            stable_count += 1
        else:
            stable_count = 0

        # If the frame is stable for more than 2 consecutive times, skip processing
        if stable_count > 2:
            continue

    # Update the previous frame
    prev_frame_gray = frame_gray.copy()

    # Use pytesseract to extract text from the frame
    text_content = pytesseract.image_to_string(frame)

    # Accumulate the text content
    all_text_content += text_content + "\n"

    # Print the frame and its text content
    print("Text content in the frame:")
    print(text_content)

    # Display the frame
    cv2.imshow("Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()

# Write all the accumulated text content to a text file
output_file = "all_text_content.txt"
with open(output_file, "w") as file:
    file.write(all_text_content)

print("All text content written to", output_file)
```

```python
# Read the contents of the text file
input_file_path = 'all_text_content.txt'
with open(input_file_path, 'r') as file:
    lines = file.readlines()

# Remove redundant lines and keep track of unique lines
unique_lines = set()
for line in lines:
    line = line.strip()  # Remove leading and trailing whitespaces
    if line:  # Check if the line is not empty after stripping
        unique_lines.add(line)

# Write the unique lines back to the text file
output_file_path = 'unique_text_file.txt'
with open(output_file_path, 'w') as file:
    for line in unique_lines:
        file.write(line + '\n')

print("Redundant text removed. Unique text saved to:", output_file_path)
```
